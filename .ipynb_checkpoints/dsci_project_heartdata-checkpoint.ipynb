```r
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/bak" 
heart_data = read_csv(url, col_names = c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num")) %>% 
    select(age, sex, trestbps, cp, restecg, num) #explain why we are using these variables 
# explain "num" is the variable we are trying to find, everything else is a predictor variable 
# explain what the values under each variable is
heart_data %>% 
    filter(trestbps != "?")
```
